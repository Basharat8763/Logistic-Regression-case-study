# 1-Statistician vs. ML Practitioner

* Definition:The approach to linear regression differs between statisticians and machine learning practitioners, largely due to their
differing goals and contexts.

### Statistician:
+ Focus: Inference and understanding relationships.
+ Objective: To interpret the coefficients and understand the underlying relationships between variables.
+ Approach: Statisticians often focus on meeting the assumptions of linear regression to ensure valid inference. They might
spend more time on exploratory data analysis (EDA) and diagnostic tests to validate these assumptions.

### ML Practitioner:
+ Focus: Prediction accuracy.
+ Objective: To build models that generalize well to new, unseen data.
+ Approach: Machine learning practitioners might be less concerned with the strict adherence to assumptions as long as the
model performs well in terms of predictive accuracy. They might use cross-validation, regularization, or more complex models
to improve predictions.

### Example:
+ Statistical Perspective: A statistician might use linear regression to understand how various factors (e.g., education level, work

# 2- Linear Regression Assumptions 
* For a linear regression model to provide valid results, certain assumptions must be met: 

## 2.2.1 Linearity
* Definition: The relationship between the independent variables and the dependent variable is linear.
* Example: In predicting house prices, the assumption is that a one-unit increase in size results in a constant change in price, irrespective of the house size.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data: House features and prices
data = {'Size (sq ft)': [1500, 1600, 1700, 1800, 1900],
        'Bedrooms': [3, 3, 3, 4, 4],
        'Price ($)': [300000, 320000, 340000, 360000, 380000]}

df = pd.DataFrame(data)

# Plotting the scatter plot to check linearity assumption
plt.scatter(df['Size (sq ft)'], df['Price ($)'])
plt.xlabel('Size (sq ft)')
plt.ylabel('Price ($)')
plt.title('Linearity Check')
plt.show()
```

## 2.2.2 Independence
* Definition: The observations in the dataset should be independent of each other.
* Example: House prices should be independent of each other, meaning the price of one house shouldn't affect the price of another. 

## 2.2.3 Homoscedasticity
* Definition: The residuals (errors) of the model should have constant variance across all levels of the independent variables.
* Example: The spread of residuals should be roughly the same for all predicted house prices.

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Features (Size) and target (Price)
X = df['Size (sq ft)'].values.reshape(-1, 1)
y = df['Price ($)'].values

# Creating and training the model
model = LinearRegression()
model.fit(X, y)

# Predicting prices based on the model
predicted_prices = model.predict(X)

# Checking homoscedasticity by plotting residuals vs. fitted values
residuals = y - predicted_prices
plt.scatter(predicted_prices, residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Homoscedasticity Check')
plt.show()
```

## 2.2.4 Normality of Residuals
* Definition: The residuals should be normally distributed.
* Example: The differences between the actual and predicted house prices should follow a normal distribution.

```python
# Checking normality of residuals using a histogram
plt.hist(residuals, bins=10, edgecolor='black')
plt.xlabel('Residuals')
plt.title('Normality Check')
plt.show() 
```

## 2.2.5 No Multicollinearity
* Definition: The independent variables should not be highly correlated with each other.
* Example: If both house size and number of rooms are included in the model, they shouldn't be strongly correlated, as this can
distort the coefficients. 


```python
# Checking multicollinearity using the correlation matrix
corr_matrix = df[['Size (sq ft)', 'Bedrooms']].corr()
print(corr_matrix) 
```

NOTE-High correlation (close to 1) between independent variables suggests multicollinearity, which could be problematic. 

# 3- Statistical Inferences from a Linear Regression Model
Statistical inferences involve drawing conclusions about the population based on the sample data and the linear regression model. Key
aspects include:
## 3.1 Coefficients
* Interpretation: The coefficients represent the change in the dependent variable for a one-unit change in the independent variable,
holding other variables constant.
* Example: In the house price model, the coefficient for "Size (sq ft)" indicates the change in price per additional square foot. 


```python
# outputting the coefficients
print(f"Intercept: {model.intercept_:.2f}")
print(f"Coefficients: {model.coef_}") 
```

## 3.2 p-values
* Definition: p-values indicate the significance of each coefficient. A low p-value (typically < 0.05) suggests that the corresponding
variable is statistically significant in predicting the dependent variable.
* Example: A low p-value for "Size (sq ft)" suggests that it is a significant predictor of house prices.

```python
 import statsmodels.api as sm
# Adding a constant for intercept
X_with_const = sm.add_constant(X)
# Fitting the model
model_sm = sm.OLS(y, X_with_const).fit()
# Summary of the model
print(model_sm.summary()) 
```

NOTE- The output includes p-values for each coefficient, helping us determine which variables significantly affect the dependent
variable. 

## 3.3 Confidence Intervals
* Definition: A confidence interval gives a range of values within which the true population parameter (e.g., a coefficient) is expected
to fall, with a certain level of confidence (usually 95%).
* Example: A 95% confidence interval for the "Size (sq ft)" coefficient might be 8070120, meaning we are 95% confident that the true
effect of size on price lies within this range. 


```python
 # Confidence intervals for the coefficients
print(model_sm.conf_int()) 
```

NOTE- These intervals provide an estimate of the uncertainty around the coefficient values.
